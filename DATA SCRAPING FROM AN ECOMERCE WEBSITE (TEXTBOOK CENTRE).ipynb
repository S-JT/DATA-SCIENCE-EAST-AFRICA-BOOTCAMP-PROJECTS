```python
#import libraries

from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
import time
import datetime
```

```python
today = datetime.date.today()
```

```python
products = []
for i in range(1,5):
  url =  f"https://textbookcentre.com/catalogue/on-sale/?sort_by=popularity&items_per_page-"
  headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}
  response = requests.get(url)
  #response = response.content
  soup = BeautifulSoup(response.text, 'html.parser')
 # soup2 =  BeautifulSoup(soup1.prettify(), "html.parser")
  #all = soup2.find_all(class_="card-title product-card-name")


  titles = soup.find_all(class_='card-title product-card-name')
  product_titles = [title.text.strip() for title in titles]
  p1 = soup(class_='stockrecord-price-current')
  product_price = [p1.text.strip() for p1 in p1]
  p2 = soup(class_='stockrecord-price-old')
  previous_price = [p2.text.strip() for p2 in p2]
  sale = soup(class_='product-ribbon-sale')
  sale_discount = [sale.text.strip() for sale in sale]

  Links = soup.find_all('a', attrs={'class':'/catalogue/lenovo-loq-3-gaming-series-i5-8gb-512gb-ssd-rtx2050-4gb-win-11-156_371085/'})
  Links_list = []
  #loop for extracting links from products
  for Link in Links:
    Links_list.append(Link.get('href'))
     #create a dictionary
  d = {"product_titles":[], "product_price":[], "previous_price":[], "sale_discount":[]}
  #loop to extract product details from each link
  for Link in Links_list:
    new_response = requests.get("https://textbookcentre.com/catalogue/category/books/?page=71" + Link, headers)
    new_soup = BeautifulSoup(new_response.content, "html.parser")

#use functions to display the necessary product information
d['product_titles'].append(product_titles)
d['product_price'].append(product_price)
d['previous_price'].append(previous_price)
d['sale_discount'].append(sale_discount)

#print(product_titles)
#print(product_price)
#print(previous_price)
#print(sale_discount )
#print(today)
```

```python
TBC_df = pd.DataFrame.from_dict(d)
TBC_df['product_titles'].replace('', np.nan,inplace=True)
TBC_df = TBC_df.dropna(subset=['product_titles'])
TBC_df.to_csv("TBC_data.csv", header=True, index=False)
```

```python
import pandas as pd
import ast

clean_df = pd.read_csv('/content/TBC_data.csv')

#initializing lists
product_titles_list = []
product_price_list = []
previous_price_list = []
sale_discount_list = []

#iterating each row
for index, row in df.iterrows():
#evaluating strings using ast
  product_titles_data = ast.literal_eval(row['product_titles'])
  product_price_data = ast.literal_eval(row['product_price'])
  previous_price_data = ast.literal_eval(row['previous_price'])
  sale_discount_data = ast.literal_eval(row['sale_discount'])

  #extending corresponding lists
  product_titles_list.extend(product_titles_data)
  product_price_list.extend(product_price_data)
  previous_price_list.extend(previous_price_data)
  sale_discount_list.extend(sale_discount_data)

  #creating a new dataframe with the split data
  new_clean_df = pd.DataFrame({'product name': product_titles_list, 'product price (KES)': product_price_list, 'previous price (KES)': previous_price_list, '% sale discount': sale_discount_list})

  print(new_clean_df)
```

```python
new_clean_df.to_csv("New clean_data.csv", header=True, index=False)
```

```python
#stripping unwanted items
with open('New clean_data.csv', 'r') as infile:
  reader = csv.reader(infile)
  header = next(reader) #this skips the header
  cleaned_product_name = []
  cleaned_product_price = []
  cleaned_previous_price = []
  cleaned_sale_discount = []

  for row in reader:
    cleaned_product_name.append
    cleaned_product_price.append(row[0].strip())
    cleaned_previous_price.append(row[0].strip())
    cleaned_sale_discount.append(row[1].strip())

with open('TBC SCRAPPER_data.csv', 'w', newline='') as outfile:
    writer = csv.writer(outfile)
    writer.writerow(header)
    for i in range(len(cleaned_product_name)):
      writer.writerow([cleaned_product_name[i], cleaned_product_price[i], cleaned_previous_price[i], cleaned_sale_discount[i]])
```
